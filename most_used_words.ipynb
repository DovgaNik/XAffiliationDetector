```python
from collections import Counter
import re
```

```python
def most_used_words(strings):
    combined_string = ' '.join(strings)
    combined_string = re.sub(r'[^a-zA-Z0-9\s]', '', combined_string.lower())
    words = combined_string.split()

    word_counts = Counter(words)
    top_10_words = word_counts.most_common(500)
    return top_10_words
```

```python
f_r = open("../../DATA/newtweets/democrat.csv", "r")
lines_r = f_r.readlines()
f_r.close()
dem_words = most_used_words(lines_r)
dem_words
```

```python
f_d = open("../../DATA/newtweets/republican.csv", "r")
lines_d = f_d.readlines()
f_d.close()
rep_words = most_used_words(lines_d)
rep_words
```

```python
dem_clean = []

for dem in dem_words:
    mentioned = False
    for rep in rep_words:
        if dem[0] == rep[0]:
            mentioned = True
    if not mentioned:
        dem_clean.append(dem)
```

```python
rep_clean = []

for rep in rep_words:
    mentioned = False
    for dem in dem_words:
        if dem[0] == rep[0]:
            mentioned = True
    if not mentioned:
        rep_clean.append(rep)
```
